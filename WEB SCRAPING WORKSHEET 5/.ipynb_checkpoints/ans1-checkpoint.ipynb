```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
wikipedia = urlopen('https://en.wikipedia.org/wiki/Main_Page')
```

```python
bs = BeautifulSoup(wikipedia, "html.parser")
titles = bs.find_all(['h1','h2','h3','h4','h5','h6'])
```

```python
print(*titles, sep='\n')
```
